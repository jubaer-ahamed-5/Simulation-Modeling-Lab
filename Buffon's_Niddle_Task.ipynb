```
import random
import math
import matplotlib.pyplot as plt
```

```
hits=0
n=2000
missD=[] 
hitD=[]
missTheta=[]
hitTheta=[]
d= int(input("Enter Value d:"))
L= int(input("Enter Value L:"))
for i in range(0,n):
  D= random.uniform(0,d/2)              
  theta= random.uniform(0,math.pi)         
  if D<=L/2*math.sin(theta):
    hits=hits+1
    hitD.append(D)
    hitTheta.append(theta)
  else:
    missD.append(D)
    missTheta.append(theta)


PI= ((2*L)/d)*(n/hits)
print(PI)
plt.scatter(missTheta,missD,color="green",label="Miss Points")
plt.scatter(hitTheta,hitD,color="red",label="Hit Points")
plt.xlabel('Value of Theta') 
plt.ylabel('Value of D')
plt.legend()
plt.show()
```
