```
Inputlist=[1,3,5,8,9,10,15,20,25,26]
even_count,odd_count=0,0
for i in Inputlist:
  if i%2==0:
    even_count+=1
    Inputlist.remove(i)
  else:
    odd_count+=1
print('Even Item: ',even_count)
print('Odd Item: ',odd_count)
print('List After Remove Even Number :',Inputlist)
```

```
line=int(input('Enter Line Size :'))
line=int(line/2)
for i in range(0,line+1):
  print()
  for j in range(0,i+1):
    print('#',end=' ')

for i in range(line+1, 0, -1):
    print()
    for j in range(0,  i-1):
        print("#", end=' ')

```

```
import matplotlib.pyplot as plt  

month_List  = [1,2,3,4,5,6]
ToothPaste_Sales_Data = [2500,2630,2140,3400,3600,2760]
FaceWash_Sales_Data   = [1500,1200,1340,1130,1740,1555]
Shampoo_Sales_Data   = [5200,5100,4550,5870,4560,4890]
Moisturizer_Sales_Data = [9200,6100,9500,8870,7760,7490]
Soap_Sales_Data   = [1200,2100,3500,1870,1560,1890]

plt.plot(month_List, ToothPaste_Sales_Data, label = 'ToothPaste')
plt.plot(month_List, FaceWash_Sales_Data,   label = 'Facewash')
plt.plot(month_List, Shampoo_Sales_Data, label = 'Shampoo')
plt.plot(month_List, Moisturizer_Sales_Data, label = 'Moisturizer')
plt.plot(month_List, Soap_Sales_Data, label = 'Soap')

plt.title('Sales data')
plt.xlabel('Month No:')
plt.ylabel('Sales unit')
plt.yticks([0, 2000, 4000, 6000, 8000, 10000, 12000, 14000])
plt.legend()
plt.show()
```
